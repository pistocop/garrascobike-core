# Thread-Product matrix EDA

```python
import pandas as pd

df_raw = pd.read_csv('../../data/03_correlation_data/presence_dataset/20210331124802/presences.csv')
df = df_raw.copy()
print(f"Dataset size:{df.memory_usage(index=True).sum()}") # dtype category = 4_209_304 | no dtype category = 2_928_976 
df_raw.head(3)
```

### Brands number distribution

```python
df_threads = df.groupby("thread_id")["product_name"].apply(list).reset_index(name="brands")
df_threads["n_brands"] = df_threads["brands"].apply(len)
df_threads = df_threads.sort_values("n_brands").reset_index()
```

```python
n_brands_mean = df_threads["n_brands"].mean()
print(f"Mean of the number of brands per thread: {n_brands_mean:.2f}")
df_threads[["n_brands"]].plot()
```

### Brands distribution

```python
df_brands = df.groupby("product_name")["thread_id"].apply(list).reset_index(name="threads")
df_brands["n_threads"] = df_brands["threads"].apply(len)
df_brands = df_brands.sort_values("n_threads").reset_index()
```

```python
n_threads_mean = df_brands["n_threads"].mean()
print(f"Mean of the number of threads per brand: {n_threads_mean:.2f}")
df_brands[["n_threads"]].plot()
```
